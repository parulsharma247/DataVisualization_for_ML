# Basic plots with Matplotlib

## Matplotlib

```python
import matplotlib.pyplot as plt
```

Q. Perform a line plot, where x axis value is(1,2,3,5) and y axis value is(2,4,1,3) **(2 mark)**

```python
## START CODE HERE : 
fig, ax= plt.subplots()
ax.plot([1,2,3,5], [2,4,1,3])
plt.show()
# plt.plot([1,2,3,5], [1,2,3,5])
## END CODE
```

## Scatter plot

Q. Perform a scatter plot, where x axis value is(10,35,15,25,50) and y axis vlaue is(1.1,3.5,1.5,2.5,5.0) **(2 mark)**

```python
## START CODE HERE : 
import numpy as np
colors = np.random.rand(5)
area = (30 * np.random.rand(5))**2

plt.scatter(x=[10,35,15,25,50], y=[1.1,3.5,1.5,2.5,5.0], s = area, c=colors)

## END CODE
```

Q. Plot a histogram, where values are (20, 45, 75, 90, 100, 105, 130, 145) and keepin the bins as 6. **(2 marks)**

```python
## START CODE HERE : 
plt.hist([20,45,75,90,100,105,130,145],bins=6)
plt.show()
## END CODE
```

Solve the following Plot line questions using matplotlib where:

Q1. x axis value is(10, 20, 30, 40) and y axis value is(20, 40, 10, 30)

Q2. Add x , y axis labels stating "x-axis" and "y-axis" respectively.**(1 mark)**

Q3. Add title as "Matplotlib". **(1 mark)**

Q4. Add double ticks in y axis with values (0, 2, 4, 6) **(1 mark)**

Q5. Add another value x = (10, 20, 30, 40) and y = (40, 10, 30, 20) and show it. **(1 mark)**

```python
## 1. START CODE HERE : 
fig, ax= plt.subplots()
ax.plot([10,20,30,40], [20,40,10,30])
## END CODE

## 2. START CODE HERE : 
ax.set_xlabel('x-axis')
ax.set_ylabel('y-axis')
## END CODE
```

```python
## 3. START CODE HERE : 
fig, ax= plt.subplots()
ax.plot([10,20,30,40], [20,40,10,30])
ax.set_xlabel('x-axis')
ax.set_ylabel('y-axis')

ax.set_title('Matplotlib')
plt.grid(True)

## END CODE## 
```

```python
## 4. START CODE HERE : 
fig, ax= plt.subplots()
ax.plot([10,20,30,40], [20,40,10,30])
ax.set_xlabel('x-axis')
ax.set_ylabel('y-axis')
ax.set_title('Matplotlib')

ax.set_yticks([0, 2, 4, 6])
plt.grid(True)
## END CODE
```
